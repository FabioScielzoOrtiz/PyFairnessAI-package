## Processing the data

```python
homecredit_data = pd.read_csv(r"C:\Users\fscielzo\Documents\IBiDat\Fairness AI\data\homecredit.csv", 
                              header=0, sep=',', quotechar='"')
homecredit_data.head()
```

### Processing without pipelines

```python
# Non-data-leakage processing

## Identifying categorical and quantitative variables

cat_cols = ['NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'NAME_TYPE_SUITE',
            'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE',
            'OCCUPATION_TYPE', 'WEEKDAY_APPR_PROCESS_START', 'ORGANIZATION_TYPE', 'FONDKAPREMONT_MODE',
            'HOUSETYPE_MODE', 'WALLSMATERIAL_MODE', 'EMERGENCYSTATE_MODE']
            # CODE_GENDER is not considered since we will use AGE as sensitive attribute


quant_cols = [x for x in homecredit_data.columns if x not in cat_cols + ['SK_ID_CURR']]
             # SK_ID_CURR is not considered as quant col

## NA values in OWN_CAR_AGE converted to 0: "no car"
homecredit_data.loc[homecredit_data['FLAG_OWN_CAR'] == "N", "OWN_CAR_AGE"] = 0

## Conversion of age from days_from_birth to age (in years)
homecredit_data['AGE'] = - homecredit_data['DAYS_BIRTH'].astype('float') / 365

## Dropping CODE_GENDER (not considered since we will use AGE as sensitive attribute), SK_ID_CURR and DAYS_BIRTH (we now have AGE)
cols_to_drop = ['DAYS_BIRTH', 'SK_ID_CURR', 'CODE_GENDER']
homecredit_data = homecredit_data.drop(columns=cols_to_drop)

## Transforming AGE into categorical (this doesn't lead to data-leakage  but must be done after imputing)
homecredit_data['AGE'] = homecredit_data['AGE'].map(lambda x : 1 if x >= 25 else 0, # AGE will be the sensitive variable and 1 (>= 25) the label of the privileged class
                                                    na_action='ignore') 

## Updating columns list after above process
quant_cols = [x for x in quant_cols if x not in cols_to_drop] 
cat_cols = [x for x in cat_cols if x not in cols_to_drop] + ['AGE']


# Data-leakage processing

## NA values converted to median in numeric variables:
homecredit_data[quant_cols] = homecredit_data[quant_cols].fillna(homecredit_data[quant_cols].median())

## NA values converted to most frequent in categorical variables
homecredit_data[cat_cols] = homecredit_data[cat_cols].fillna(homecredit_data[cat_cols].mode().iloc[0])

## OneHot encoding the categorical variables (this doesn't lead to data-leakage necessarily but must be done after imputing)
encoder_ = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first')
onehotencoding_arr = encoder_.fit_transform(homecredit_data[cat_cols])
onehotencoding_df = pd.DataFrame(onehotencoding_arr, columns=encoder_.get_feature_names_out())
homecredit_data = pd.concat([homecredit_data[quant_cols], onehotencoding_df], axis=1)
```

### Processing with pipelines

```python
homecredit_data = pd.read_csv(r"C:\Users\fscielzo\Documents\IBiDat\Fairness AI\data\homecredit.csv", header=0, sep=',', quotechar='"')
homecredit_data.head()
```

```python
# Non-data-leakage processing

## Identifying categorical and quantitative variables

cat_cols = ['NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'NAME_TYPE_SUITE',
            'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE',
            'OCCUPATION_TYPE', 'WEEKDAY_APPR_PROCESS_START', 'ORGANIZATION_TYPE', 'FONDKAPREMONT_MODE',
            'HOUSETYPE_MODE', 'WALLSMATERIAL_MODE', 'EMERGENCYSTATE_MODE']
            # CODE_GENDER is not considered since we will use AGE as sensitive attribute


quant_cols = [x for x in homecredit_data.columns if x not in cat_cols + ['SK_ID_CURR']]
             # SK_ID_CURR is not considered as quant col

## NA values in OWN_CAR_AGE converted to 0: "no car"
homecredit_data.loc[homecredit_data['FLAG_OWN_CAR'] == "N", "OWN_CAR_AGE"] = 0

## Conversion of age from days_from_birth to age (in years)
homecredit_data['AGE'] = - homecredit_data['DAYS_BIRTH'].astype('float') / 365

## Dropping CODE_GENDER (not considered since we will use AGE as sensitive attribute), SK_ID_CURR and DAYS_BIRTH (we now have AGE)
cols_to_drop = ['DAYS_BIRTH', 'SK_ID_CURR', 'CODE_GENDER']
homecredit_data = homecredit_data.drop(columns=cols_to_drop)

## Transforming AGE into categorical (this doesn't lead to data-leakage  but must be done after imputing)
homecredit_data['AGE'] = homecredit_data['AGE'].map(lambda x : 1 if x >= 25 else 0, # AGE will be the sensitive variable and 1 (>= 25) the label of the privileged class
                                                    na_action='ignore') 

## Updating columns list after above process
quant_cols = [x for x in quant_cols if x not in cols_to_drop] 
cat_cols = [x for x in cat_cols if x not in cols_to_drop] + ['AGE']
```

```python
quant_pipeline = Pipeline([
    ('imputer', imputer(apply=True, method='simple_median'))
    ])

cat_pipeline = Pipeline([
    ('imputer', imputer(apply=True, method='simple_most_frequent')),
    ('encoder', encoder(method='one-hot', drop='first'))
    ])

quant_cat_processing = ColumnTransformer(transformers=[('quant', quant_pipeline, quant_cols),
                                                       ('cat', cat_pipeline, cat_cols)])
```

```python
processed_data = quant_cat_processing.fit_transform(X=homecredit_data)
# Updating cat cols
cat_cols = quant_cat_processing.named_transformers_['cat']['encoder'].get_feature_names_out(cat_cols).tolist()
# Transforming 'processed_data' into Pandas df
homecredit_data = pd.DataFrame(processed_data, columns = quant_cols + cat_cols)
```
